```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Ex1

```python
file = 'Data_EX1.txt'
with open(file , 'r') as f:
    lines = f.readlines()

lines
```

```python
header = lines[0].strip().split()
header = [header[0]] + [str(i) for i in range (1,13)]

data = [line.strip().split() for line in lines[0:]]

df = pd.DataFrame(data, columns = header)
df = df.dropna()
df
```

```python
columns = df.iloc[:, 1 :]
columns.replace(str(-99.0), np.nan, inplace = True)
columns.replace(str(99.0), np.nan, inplace = True)
columns = columns.apply(pd.to_numeric, errors ='coerce')
columns
```

```python
data_value = []
for col in columns:
    data_value.append(columns[col].mean())  

# Danh sch thng
time = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# V biu 
plt.figure(figsize=(15, 5))

# V nhit  hng ngy
daily_temps = columns.values.T  
x_days = np.arange(1, len(columns.values.flatten()) + 1)  # Cc ngy trong nm
plt.plot(x_days, daily_temps.flatten(), color='blue', alpha=0.5, label='Daily temperature')

# V nhit  trung bnh hng thng
monthly_positions = np.linspace(1, len(x_days), len(data_value))  # V tr trung bnh thng
plt.plot(monthly_positions, data_value, linestyle='-', color='red', marker='o',markerfacecolor='green',markeredgecolor='green', markersize=8, label='Monthly mean temperature')


# Ghi ch gi tr trung bnh hng thng
for i in range(len(time)):
    if i < 3 or i >= len(time) - 1:
        plt.text(monthly_positions[i], data_value[i] - 1, f'{data_value[i]:.2f}', ha='center', va='top', color='green')
    else:
        plt.text(monthly_positions[i], data_value[i] + 1, f'{data_value[i]:.2f}', ha='center', va='bottom', color='green')

# Ty chnh trc x
plt.xticks(monthly_positions, time)  # Gn nhn trc x theo thng

# Ci t trc y
plt.ylim(5, 35)

# Thm cc thng tin vo biu 
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.title('Daily and Monthly Mean Temperature in 2009')
plt.legend()
plt.grid()


plt.show()
```
