```python
file = 'Data_EX1.txt'
with open(file , 'r') as f:
    lines = f.readlines()

lines
```

```python
header = lines[0].strip().split()
header = [header[0]] + [str(i) for i in range (1,13)]

data = [line.strip().split() for line in lines[0:]]

df = pd.DataFrame(data, columns = header)
df = df.dropna()
df
```

```python
columns = df.iloc[:, 1 :]
columns.replace(str(-99.0), np.nan, inplace = True)
columns.replace(str(99.0), np.nan, inplace = True)
columns = columns.apply(pd.to_numeric, errors ='coerce')
columns
```

```python
data_value = []
for col in columns:
    data_value.append(columns[col].mean())  

# Danh sch thng
time = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

# V biu 
plt.figure(figsize=(15, 5))

# V nhit  hng ngy
daily_temps = columns.values.T  
x_days = np.arange(1, len(columns.values.flatten()) + 1)  # Cc ngy trong nm
plt.plot(x_days, daily_temps.flatten(), color='blue', alpha=0.5, label='Daily temperature')

# V nhit  trung bnh hng thng
monthly_positions = np.linspace(1, len(x_days), len(data_value))  # V tr trung bnh thng
plt.plot(monthly_positions, data_value, linestyle='-', color='red', marker='o',markerfacecolor='green',markeredgecolor='green', markersize=8, label='Monthly mean temperature')


# Ghi ch gi tr trung bnh hng thng
for i in range(len(time)):
    if i < 3 or i >= len(time) - 1:
        plt.text(
            monthly_positions[i], 
            data_value[i] + 1.5,  # y s liu ln trn cao hn
            f'{data_value[i]:.2f}', 
            ha='center', 
            va='bottom',  # Cn chnh di cng ca ch,  ch khng b ct
            color='black',
            fontsize=10
        )
    else:
        plt.text(
            monthly_positions[i], 
            data_value[i] + 1.5,  # y s liu ln trn cao hn
            f'{data_value[i]:.2f}', 
            ha='center', 
            va='bottom',  # Cn chnh di cng ca ch
            color='black',
            fontsize=10
        )
# Ty chnh trc x
plt.xticks(monthly_positions, time)  # Gn nhn trc x theo thng

# Ci t trc y
plt.ylim(5, 35)

# Thm cc thng tin vo biu 
plt.xlabel('Month')
plt.ylabel('Temperature (C)')
plt.title('Daily and Monthly Mean Temperature in 2009')
plt.legend()
plt.grid()


plt.show()
```

```python
from numpy import *

import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
from matplotlib.lines import Line2D
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.colors as colors
from matplotlib.colors import BoundaryNorm
from matplotlib.patches import Rectangle


# Set the random seed for reproducibility
random.seed(180793)

# Generate random sales data for 10 stores and 12 months (for 3 fruits)
apple = random.randint(0, 30, size=(10, 12))  # Apple sales data
banana = random.randint(0, 60, size=(10, 12))  # Banana sales data
durian = random.randint(0, 80, size=(10, 12))  # Durian sales data

# Calculate total sales for each store (sum across months)
total_sales_per_store = apple.sum(axis=1) + banana.sum(axis=1) + durian.sum(axis=1)

# Calculate total sales for each month (sum across stores)
total_sales_per_month = apple.sum(axis=0) + banana.sum(axis=0) + durian.sum(axis=0)

# Function to calculate radius for the size of pies
def calculate_radius(sales):
    if sales < 40:
        return 0.1
    elif sales < 60:
        return 0.3
    elif sales < 80:
        return 0.5
    elif sales < 100:
        return 0.7
    else:
        return 1

# Total sales data for all stores and months
total_sales = apple + banana + durian

# Set up the plot
fig, ax = plt.subplots(figsize=(15, 10))

# We will create a grid for 10 stores (rows) and 12 months (columns)
# Plot pie charts for each month-store combination
for i in range(10):  # For each store
    for j in range(12):  # For each month
        # Calculate the sales for the current store and month
        sales_apple = apple[i, j]
        sales_banana = banana[i, j]
        sales_durian = durian[i, j]
        
        # Calculate the size of the pie chart based on the total sales
        total_sales_this_month = sales_apple + sales_banana + sales_durian
        radius = calculate_radius(total_sales_this_month)
        
        # Plot a pie chart at the appropriate location (month, store)
        ax.pie([sales_apple, sales_banana, sales_durian], 
               labels=None,  # Don't show labels
               startangle=90, 
               radius=radius,
               colors=['orange', '#1E90FF', 'forestgreen'],  # Colors for apple, banana (blue), durian (green)
               center=(j*2 + 1, (9-i)*2 + 1))  # Adjust center position for each pie chart (reverse y-axis)

# Set axis labels and ticks
ax.set_xticks(np.arange(0, 12 * 2, 2))
ax.set_xticklabels(["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"])

ax.set_yticks(np.arange(0, 10 * 2, 2))
ax.set_yticklabels([f"Store {i+1}" for i in range(9, -1, -1)])  # Reverse the store labels (Store 10 to Store 1)

# Add gridlines for the Y-axis (black lines separating the rows)
# Plot horizontal lines between each store (draw lines between rows)
for i in range(1, 12):
    ax.plot([0, 12*2], [(i*2), (i*2)], color='black', lw=1)

# Add borders around the Y-axis labels (mimicking pie chart borders)
for i in range(10):
    ax.plot([-1, 13*2], [(i*2), (i*2)], color='black', lw=1, zorder=1)  # Upper border for each store row
    ax.plot([-1, 13*2], [((i+1)*2), ((i+1)*2)], color='black', lw=1, zorder= 1)  # Lower border for each store row
handles = [
    Rectangle((0, 0), 1, 0.4, color='orange', label='apple'),        # Rectangle for Apple
    Rectangle((0, 0), 1, 0.4, color='#1E90FF', label='banana'),      # Rectangle for Banana
    Rectangle((0, 0), 1, 0.4, color='forestgreen', label='durian')   # Rectangle for Durian
]

# Add the legend to the plot and position it at the top right, just above Store 1
ax.legend(handles=handles, loc='upper left', bbox_to_anchor=(0.7, 1.2), ncol=3)
# Show the plot
plt.tight_layout()
plt.show()
```

```python
file2 = 'Data_EX6.txt'
with open(file2 , 'r') as f:
    lines2 = f.readlines()

lines2
```

```python
header2 = lines2[0].strip().split()
header2 = [header2[0]] + [str(i) for i in range (1,13)]

data2 = [line.strip().split() for line in lines2[0:]]

df2 = pd.DataFrame(data2, columns = header2)
df2 = df2.dropna()
df2
```

```python
df2 = df2.iloc[:, 1 :]
df2.replace(str(-99.0), np.nan, inplace = True)
df2.replace(str(99.0), np.nan, inplace = True)
df2 = df2.apply(pd.to_numeric, errors ='coerce')
df2
```

```python
def assign_season(month):
    if month in [12, 1, 2]:
        return 'DJF'
    elif month in [3, 4, 5]:
        return 'MAM'
    elif month in [6, 7, 8]:
        return 'JJA'
    else:
        return 'SON'

# Ly ct 'month' t header  gn ma
months = [int(m) for m in header2[1:]]
seasons = [assign_season(month) for month in months]
df2['season'] = [seasons[i % 12] for i in range(len(df2))]
df2['observation'] = df2.iloc[:, 0]
for col in df2.columns[1:]:
    df2[col] = df2.apply(lambda row: row[col] if not np.isnan(row['observation']) else np.nan, axis=1)

# Tnh tn sut theo ngng (0, 5, 10, 15, 20, 25, 30 mm)
df2['observation'] = df2.iloc[:, 0]
for col in df2.columns[1:]:
    df2[col] = df2.apply(lambda row: row[col] if not np.isnan(row['observation']) else np.nan, axis=1)

# Tnh tn sut theo ngng (0, 5, 10, 15, 20, 25, 30 mm)
thresholds = [0, 5, 10, 15, 20, 25, 30]
seasons = ['DJF', 'MAM', 'JJA', 'SON']

# V biu  tn sut cho tng ma
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, season in enumerate(seasons):
    seasonal_data = df2[df2['season'] == season]
    
    freq_obs = [(seasonal_data['observation'] >= t).sum() for t in thresholds]
    freq_simulations = {}
    
    for col in df2.columns[1:]:
        if col != 'season' and col != 'observation':
            freq_simulations[col] = [(seasonal_data[col] >= t).sum() for t in thresholds]
    
    # V d liu quan st
    axes[i].bar(thresholds, freq_obs, width=3, label='Observation', alpha=0.6)
    
    # V d liu m phng
    for sim, freq in freq_simulations.items():
        axes[i].bar(thresholds, freq, width=3, label=sim, alpha=0.6)
    
    axes[i].set_title(f'Precipitation Frequency by Threshold ({season})')
    axes[i].set_xlabel('Rain Amount (mm)')  # i tn trc X thnh "Rain Amount (mm)"
    axes[i].set_ylabel('Frequency')
    axes[i].legend()

plt.tight_layout()
plt.show()
```
